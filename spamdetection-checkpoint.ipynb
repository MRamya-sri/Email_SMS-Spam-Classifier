3.DATA PREPROCESSING
-> Lower Case
-> Tokenization
-> Removing special characters
-> Removing stop words and punctuation
-> Stemming

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stopwords.words('english')
```

```python
import string 
string.punctuation
```

```python
from nltk.stem.porter import PorterStemmer
ps =PorterStemmer()
ps.stem('loving')
```

```python
def transform_text(text):
    text = text.lower()#lowercase
    text = nltk.word_tokenize(text)#tokenization
    
    y=[]
    for i in text:
        if i.isalnum(): #removing speacial characters
            y.append(i)
    text = y[:]
    y.clear()
    
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
            
    text = y[:]
    y.clear()
    
    for i in text: #stemming
        y.append(ps.stem(i))
        
    return " ".join(y)
    
    
```

4.MODEL BULDING

```python
#using naive bayes
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
cv = CountVectorizer()
tfidf = TfidfVectorizer(max_features= 3000)
```

```python
X = tfidf.fit_transform(df['transformed_text']).toarray()
```

```python
X
```

```python
X.shape
```

```python
y = df['target'].values
```

```python
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2,random_state=2)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
gnb = GaussianNB() #continuous values
mnb = MultinomialNB()#discrete
bnb = BernoulliNB() #binary,boolean like yes/or, 0/1
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
```

->Trying gnb

```python
gnb.fit(X_train,y_train)
y_pred1 = gnb.predict(X_test)
```

```python
accuracy_score(y_test, y_pred1)
```

```python
confusion_matrix(y_test, y_pred1)
```

```python
precision_score(y_test, y_pred1)
```

Precision is low for gnb, lets try mnb

```python
mnb.fit(X_train,y_train)
y_pred2 = mnb.predict(X_test)
print(accuracy_score(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
print(precision_score(y_test,y_pred2))
```

precision score is good isn't, but lets try bnb


```python
bnb.fit(X_train,y_train)
y_pred3 = bnb.predict(X_test)
print(accuracy_score(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
print(precision_score(y_test,y_pred3))
```

mnb and bnd have slight accuracy difference but precsion score is perfect in mnb so lets take mnb for further classification.
